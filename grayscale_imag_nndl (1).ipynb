```python
import cv2
import numpy as np
from google.colab import files
import matplotlib.pyplot as plt

# Open file chooser using Colab's file upload feature
uploaded = files.upload()

if not uploaded:
    print("No file selected.")
else:
    # Assuming only one file is uploaded, get its filename and content
    filename = next(iter(uploaded))
    image_bytes = uploaded[filename]

    # Read image from bytes using numpy and cv2
    image = cv2.imdecode(np.frombuffer(image_bytes, np.uint8), cv2.IMREAD_COLOR)

    if image is None:
        print("Error loading image.")
    else:
        # Split channels (OpenCV uses BGR)
        B, G, R = cv2.split(image)

        # Print matrices
        print("Red Channel Matrix:\n")
        print(R)

        print("\nGreen Channel Matrix:\n")
        print(G)

        print("\nBlue Channel Matrix:\n")
        print(B)

        # Display results using matplotlib
        plt.figure(figsize=(15, 5))

        plt.subplot(1, 4, 1)
        # Convert BGR to RGB for matplotlib display
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.title("Original Image")
        plt.axis('off')

        plt.subplot(1, 4, 2)
        plt.imshow(R, cmap='Reds') # Using Red colormap for Red channel
        plt.title("Red Channel")
        plt.axis('off')

        plt.subplot(1, 4, 3)
        plt.imshow(G, cmap='Greens') # Using Green colormap for Green channel
        plt.title("Green Channel")
        plt.axis('off')

        plt.subplot(1, 4, 4)
        plt.imshow(B, cmap='Blues') # Using Blue colormap for Blue channel
        plt.title("Blue Channel")
        plt.axis('off')

        plt.tight_layout()
        plt.show()
```

```python
import cv2
import numpy as np
from google.colab import files
import matplotlib.pyplot as plt

# Open file chooser using Colab's file upload feature
uploaded = files.upload()

if not uploaded:
    print("No file selected.")
else:
    # Assuming only one file is uploaded, get its filename and content
    filename = next(iter(uploaded))
    image_bytes = uploaded[filename]

    # Read image from bytes using numpy and cv2
    image = cv2.imdecode(np.frombuffer(image_bytes, np.uint8), cv2.IMREAD_COLOR)

    if image is None:
        print("Error loading image.")
    else:
        # Convert to grayscale
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # 1. Blur Image
        blur = cv2.GaussianBlur(gray, (7, 7), 0)

        # 2. Left Sobel (detect vertical edges)
        sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
        sobel_x = cv2.convertScaleAbs(sobel_x)

        # 3. Top Sobel (detect horizontal edges)
        sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
        sobel_y = cv2.convertScaleAbs(sobel_y)

        # 4. Outline Filter (Canny Edge Detection)
        outline = cv2.Canny(gray, 100, 200)

        # Display results using matplotlib
        plt.figure(figsize=(15, 8))

        plt.subplot(2, 3, 1)
        # Convert BGR to RGB for matplotlib display
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.title("Original Image")
        plt.axis('off')

        plt.subplot(2, 3, 2)
        plt.imshow(blur, cmap='gray')
        plt.title("Blur Image")
        plt.axis('off')

        plt.subplot(2, 3, 3)
        plt.imshow(sobel_x, cmap='gray')
        plt.title("Left Sobel (X)")
        plt.axis('off')

        plt.subplot(2, 3, 4)
        plt.imshow(sobel_y, cmap='gray')
        plt.title("Top Sobel (Y)")
        plt.axis('off')

        plt.subplot(2, 3, 5)
        plt.imshow(outline, cmap='gray')
        plt.title("Outline Filter")
        plt.axis('off')

        plt.tight_layout()
        plt.show()
```
