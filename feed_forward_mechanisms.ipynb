feed forward using pytorch


```python
import torch
import torch.nn as nn
import torch.optim as optim

x = torch.tensor([0.5, 0.2, 0.1], dtype=torch.float32)
y = torch.tensor([1.0], dtype=torch.float32)

model = nn.Sequential(
    nn.Linear(3, 2),
    nn.Sigmoid(),
    nn.Linear(2, 1),
    nn.Sigmoid()
)

with torch.no_grad():
    model[0].weight = nn.Parameter(torch.tensor([
        [0.1, 0.2, 0.3],
        [0.4, 0.3, 0.6]
    ]))
    model[0].bias = nn.Parameter(torch.tensor([0.01, 0.02]))

    model[2].weight = nn.Parameter(torch.tensor([[0.2, 0.3]]))
    model[2].bias = nn.Parameter(torch.tensor([0.05]))

output = model(x)
print("Final Output:", output.item())

criterion = nn.MSELoss()
loss = criterion(output, y)
print("Loss:", loss.item())

optimizer = optim.SGD(model.parameters(), lr=0.1)
optimizer.zero_grad()
loss.backward()
optimizer.step()

print("Updated w2:", model[2].weight.data)
```

feed forward using tensorflow


```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x=np.array([[0.5,0.1,0.7]])
y=np.array([[0.6]])

model=Sequential([
    Dense(2,activation='sigmoid',input_shape=(3,)),
    Dense(1,activation='sigmoid')
])

model.compile(loss='mse',optimizer='sgd')

model.fit(x,y,epochs=1000,verbose=0)

output=model.predict(x)
print("output",output)
```

feed forward using class numpy

```python
import numpy as np
class NeuralNet:
    def __init__(self,input_size,hidden_size,output_size,l=0.1):
        self.l=l
        self.W=np.random.randn(input_size,hidden_size)
        self.b=np.zeros(hidden_size)
        self.W2=np.random.randn(hidden_size,output_size)
        self.b2=np.zeros(output_size)
    def sigmoid(self,x):return 1/(1+np.exp(-x))
    def sigmoid_derivative(self,x):s=self.sigmoid(x);return s*(1-s)
    def forward(self,x):
        self.z=np.dot(x,self.W)+self.b
        self.a=self.sigmoid(self.z)
        self.z2=np.dot(self.a,self.W2)+self.b2
        self.a2=self.sigmoid(self.z2)
        return self.a2
    def backward(self,x,y):
        dL_da2=-2*(y-self.a2)
        da2_dz2=self.sigmoid_derivative(self.z2)

        dL_dW2=np.outer(self.a, dL_da2*da2_dz2)
        dL_db2=dL_da2*da2_dz2

        dL_da=np.dot(self.W2, dL_da2*da2_dz2)

        da_dz=self.sigmoid_derivative(self.z)
        dL_dW=np.outer(x,dL_da*da_dz)
        dL_db=dL_da*da_dz
        self.W2-=self.l*dL_dW2
        self.b2-=self.l*dL_db2
        self.W-=self.l*dL_dW
        self.b-=self.l*dL_db
    def train(self,x,y,epochs):
        for _ in range(epochs):
            self.forward(x)
            self.backward(x,y)

x=np.array([0.5,0.1,0.7])
y=0.6
nn=NeuralNet(3,2,1,0.1)
nn.train(x,y,1000)
print("output",nn.forward(x))
```
