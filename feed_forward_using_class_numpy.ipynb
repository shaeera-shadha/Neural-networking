```python
import numpy as np
class NeuralNet:
    def __init__(self,input_size,hidden_size,output_size,l=0.1):
        self.l=l
        self.W=np.random.randn(input_size,hidden_size)
        self.b=np.zeros(hidden_size)
        self.W2=np.random.randn(hidden_size,output_size)
        self.b2=np.zeros(output_size)
    def sigmoid(self,x):return 1/(1+np.exp(-x))
    def sigmoid_derivative(self,x):s=self.sigmoid(x);return s*(1-s)
    def forward(self,x):
        self.z=np.dot(x,self.W)+self.b
        self.a=self.sigmoid(self.z)
        self.z2=np.dot(self.a,self.W2)+self.b2
        self.a2=self.sigmoid(self.z2)
        return self.a2
    def backward(self,x,y):
        dL_da2=-2*(y-self.a2)
        da2_dz2=self.sigmoid_derivative(self.z2)

        dL_dW2=np.outer(self.a, dL_da2*da2_dz2)
        dL_db2=dL_da2*da2_dz2

        dL_da=np.dot(self.W2, dL_da2*da2_dz2)

        da_dz=self.sigmoid_derivative(self.z)
        dL_dW=np.outer(x,dL_da*da_dz)
        dL_db=dL_da*da_dz
        self.W2-=self.l*dL_dW2
        self.b2-=self.l*dL_db2
        self.W-=self.l*dL_dW
        self.b-=self.l*dL_db
    def train(self,x,y,epochs):
        for _ in range(epochs):
            self.forward(x)
            self.backward(x,y)

x=np.array([0.5,0.1,0.7])
y=0.6
nn=NeuralNet(3,2,1,0.1)
nn.train(x,y,1000)
print("output",nn.forward(x))
```
